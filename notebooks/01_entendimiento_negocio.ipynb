# Predicción y análisis de factores asociados a la depresión en estudiantes

## Contexto

La salud mental estudiantil es un problema relevante en entornos académicos, ya que factores como presión académica, hábitos de sueño, alimentación, estrés financiero y antecedentes familiares pueden influir en el riesgo de depresión. Este proyecto busca usar técnicas de análisis de datos y aprendizaje de máquina para entender esas relaciones y construir un modelo predictivo útil para apoyar procesos de detección temprana.

## Problema

Las instituciones educativas pueden tener dificultades para identificar de manera temprana perfiles estudiantiles con mayor riesgo de depresión. Contar con un modelo que ayude a reconocer patrones asociados al riesgo permitiría orientar mejor acciones preventivas y de acompañamiento.

## Pregunta de negocio

¿Qué variables académicas, personales y de estilo de vida se asocian con la depresión en estudiantes, y qué tan bien puede un modelo de clasificación identificar estudiantes con mayor riesgo?

## Objetivo general

Desarrollar un proyecto de aprendizaje de máquina, bajo la metodología CRISP-DM, que permita analizar los factores asociados a la depresión en estudiantes y construir un modelo de clasificación para predecir la variable Depression.

## Objetivos específicos
Comprender la estructura y calidad del dataset raw.csv.
Identificar relaciones entre variables como presión académica, sueño, hábitos alimenticios, horas de estudio/trabajo, estrés financiero y depresión.
Preparar los datos mediante limpieza, codificación, transformación y selección de variables.
Entrenar y comparar varios modelos de clasificación.
Evaluar el desempeño del mejor modelo con métricas adecuadas para clasificación.
Comunicar hallazgos mediante un dashboard interactivo.

Desde modelado

## Riesgos y consideraciones éticas

Este dataset contiene variables sensibles. El modelo no debe presentarse como diagnóstico clínico, sino como una herramienta analítica de apoyo para identificar patrones y posibles perfiles de riesgo. Las conclusiones deben comunicarse con cuidado y sin estigmatizar.
